```python
class Node:
    def __init__(self, data):
      self.small = None
      self.big = None
      self.too_big = None
      self.data = data

    def insert(self, data):
# Compare the new value with the parent node
      if self.data:
        if data < self.data:
            if self.small is None:
               self.small = Node(data)
            else:
               self.small.insert(data)
        elif data < self.data + 10:
            if self.big is None:
                self.big = Node(data)
            else:
                self.big.insert(data)
        else:
            if self.too_big is None:
                self.too_big = Node(data)
            else:
                self.too_big.insert(data)
      else:
        self.data = data

# Print the tree
    def traversal(self):
        if self.small:
            self.small.traversal()
        print(self.data)
        if self.big:
            self.big.traversal()
        if self.too_big:
            self.too_big.traversal()

#delete node
    def delete(self, value):
        temp = self.list_out()
        if value in temp:
            temp.remove(value)
        else:
            print("Value not in tree")
        self.small = None
        self.big = None
        self.too_big = None
        self.data = None    
        for i in range(len(temp)):
            self.insert(temp[i])

    def list_out(self):
        temp = []
        if self.small:
            temp = self.small.list_out()
        temp.append(self.data)
        if self.big:
            temp = temp + self.big.list_out()
        if self.too_big:
            temp = temp +self.too_big.list_out()
        return temp
        
            

# Use the insert method to add nodes
#root = Node(12)
#root.insert(6)
#root.insert(14)
#root.insert(3)

#          10
#        /    \
#       34      89
#     /    \  /    \
#  20     45  56    54

root = Node(27)
root.insert(14)
root.insert(35)
root.insert(10)
root.insert(19)
root.insert(31)
root.insert(42)


print("In order Traversal:")
root.traversal()

root.delete(27)
print("In order Traversal:")
root.traversal()




```
